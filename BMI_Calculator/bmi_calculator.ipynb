```python
import tkinter as tk
from tkinter import messagebox

# Dictionary to store user BMI history (optional for in-memory tracking)
user_bmi_history = {}

# Function to write user data to a file
def save_to_file(name, weight, height_feet, height_inches, bmi):
    try:
        with open("bmi_data.txt", "a") as file:
            file.write(f"Name: {name}, Weight: {weight} kg, Height: {height_feet}ft {height_inches}IN, BMI: {bmi}\n")
        messagebox.showinfo("Success", "Data saved to file.")
    except Exception as e:
        messagebox.showerror("Error", f"Could not save data: {e}")

# Function to calculate BMI and store the history
def calculate_bmi():
    try:
        name = entry_name.get().strip()
        weight = float(entry_weight.get())
        feet = scale_feet.get()  # Get the value from the feet slider
        inches = scale_inches.get()  # Get the value from the inches slider
        
        # Convert feet and inches to meters
        total_height_inches = feet * 12 + inches
        height_in_meters = total_height_inches * 0.0254  # 1 inch = 0.0254 meters
        
        bmi = weight / (height_in_meters ** 2)
        bmi = round(bmi, 2)
        
        # Store the BMI in the user's history (optional)
        if name:
            if name not in user_bmi_history:
                user_bmi_history[name] = []
            user_bmi_history[name].append(bmi)
        
            # Display BMI
            label_result.config(text=f"BMI: {bmi}")

            # Save the data to a text file
            save_to_file(name, weight, feet, inches, bmi)

            # Show user's BMI history
            bmi_history = ", ".join(map(str, user_bmi_history[name]))
            label_history.config(text=f"History for {name}: {bmi_history}")
        
        else:
            messagebox.showerror("Input Error", "Please enter your name.")

        # Display a message based on BMI value
        if bmi < 18.5:
            messagebox.showinfo("BMI Result", f"Your BMI is {bmi}. You are underweight.")
        elif 18.5 <= bmi < 24.9:
            messagebox.showinfo("BMI Result", f"Your BMI is {bmi}. You have a normal weight.")
        elif 25 <= bmi < 29.9:
            messagebox.showinfo("BMI Result", f"Your BMI is {bmi}. You are overweight.")
        else:
            messagebox.showinfo("BMI Result", f"Your BMI is {bmi}. You are obese.")
    except ValueError:
        messagebox.showerror("Input Error", "Please enter valid numeric values.")

# Create the main window
root = tk.Tk()
root.title("BMI Calculator with Data Storage")
root.geometry("400x400")  # Set window size

# Create labels and entry fields for name
label_name = tk.Label(root, text="Enter your name:")
label_name.pack(pady=10)
entry_name = tk.Entry(root)
entry_name.pack()

# Create labels and entry fields for weight
label_weight = tk.Label(root, text="Enter your weight (kg):")
label_weight.pack(pady=10)
entry_weight = tk.Entry(root)
entry_weight.pack()

# Create sliders (scales) for height (feet and inches)
label_feet = tk.Label(root, text="Select your height (feet):")
label_feet.pack(pady=10)
scale_feet = tk.Scale(root, from_=1, to=7, orient=tk.HORIZONTAL)  # Slider for feet (range 1 to 7)
scale_feet.pack()

label_inches = tk.Label(root, text="Select your height (inches):")
label_inches.pack(pady=10)
scale_inches = tk.Scale(root, from_=0, to=11, orient=tk.HORIZONTAL)  # Slider for inches (range 0 to 11)
scale_inches.pack()

# Create a button to calculate BMI
button_calculate = tk.Button(root, text="Calculate BMI", command=calculate_bmi)
button_calculate.pack(pady=10)

# Create a label to display the result
label_result = tk.Label(root, text="Your BMI will be displayed here")
label_result.pack(pady=20)

# Create a label to display the BMI history
label_history = tk.Label(root, text="Your BMI history will be displayed here")
label_history.pack(pady=20)

# Run the main event loop
root.mainloop()
```
